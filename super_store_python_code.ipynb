```python
orders_df['Product Base Margin'].fillna(0, inplace=True)
```

```python
import pandas as pd

orders_and_manager_df = pd.merge(orders_df, users_df[['Region', 'Manager']], on='Region', how='left')

print(orders_and_manager_df[['Order ID', 'Region', 'Manager']].head())


```

```python
import pandas as pd

orders_and_manager_df['Order Duration'] = (orders_and_manager_df['Ship Date'] - orders_and_manager_df['Order Date']).dt.days

print(orders_and_manager_df[['Order ID', 'Order Date', 'Ship Date', 'Order Duration']].head())
```

```python
orders_and_manager_df.to_csv('updated_sales.csv', index=False)
```

```python

returns_updated_df = pd.merge(orders_and_manager_df, returns_df, on='Order ID', how='inner')



print(returns_updated_df[['Order ID', 'Region', 'State or Province', 'Manager', 'Status']].head())
returns_updated_df.to_csv('returns_sheet.csv', index=False)
```

```python
import pandas as pd


updated_orders_and_returns_df = pd.merge(orders_and_manager_df, returns_df, on='Order ID', how='left')

updated_orders_and_returns_df = updated_orders_and_returns_df.fillna('Successful')

updated_orders_and_returns_df.head()
```

```python
import pandas as pd

def gain_loss(profit):
    if profit < 0:
        return 'Loss'
    elif profit == 0:
        return 'None'
    else:
        return 'Gain'

updated_orders_and_returns_df['Gain/Loss'] = updated_orders_and_returns_df['Profit'].apply(gain_loss)
                                                                                    
updated_orders_and_returns_df.head()
```

```python
updated_orders_and_returns_df.to_csv('final_sales_sheet.csv', index=False)
```

```python
successful_orders = updated_orders_and_returns_df[updated_orders_and_returns_df['Status'] == 'Successful']

print(successful_orders)
```

```python
successful_orders.to_csv('successful_orders.csv', index=False)
```
